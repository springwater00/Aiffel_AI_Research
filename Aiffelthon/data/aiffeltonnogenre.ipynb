```python
# CSV to JSON

import pandas as pd
import json
import os

# CSV 파일 읽기
file_name = "output5.csv"
df = pd.read_csv(file_name)

# 인덱스를 사용하여 각 행을 JSON 파일로 저장
for index, row in df.iterrows():
    row_data = {key: [value] for key, value in row.items()}  # 각 행을 dictionary로 변환하고 값들을 list로 감싸줌
    a_value = row[df.columns[0]]  # A 열의 값을 가져옵니다.
    a_value_without_ext = os.path.splitext(a_value)[0]  # ".wav" 제거
    json_file_name = f"{a_value_without_ext}.json"  # 파일 이름으로 사용

    # JSON 파일 저장
    with open(json_file_name, 'w', encoding='utf-8') as json_file:
        json.dump(row_data, json_file, ensure_ascii=False, indent=4)
```

# data for fine-tuning

```python
# 전체 샘플에서 disco/hiphop/pop 장르 노래들 100개씩, sc비율 같게 추출

import pandas as pd
import numpy as np

# Excel 파일 읽기
file_path = "/Users/butterflyeffectleedea/desktop/git/Aiffel/aiffelton4.xlsx"
df = pd.read_excel(file_path)

# 각 장르별로 sc 비율이 같게 100개씩 랜덤 추출
genres = ['disco', 'hiphop', 'pop']
sc_values = ['bass', 'lead', 'pluck', 'kick', 'cymbals', 'tom', 'snare', 'hihat']
num_samples = 100
result = []

for genre in genres:
    genre_df = df[df['genre'] == genre]
    genre_samples = []
    
    # sc 비율을 유지하며 각 장르별로 추출
    for sc in sc_values:
        sc_df = genre_df[genre_df['sc'] == sc]
        sample_size = min(len(sc_df), num_samples // len(sc_values))
        genre_samples.append(sc_df.sample(n=sample_size, replace=False, random_state=42))

    genre_result = pd.concat(genre_samples)
    result.append(genre_result)

# 결과 DataFrame 생성
result_df = pd.concat(result)

# 새로운 Excel 파일로 저장
result_df.to_excel("/Users/butterflyeffectleedea/desktop/git/Aiffel/aiffelton300.xlsx", index=False)
```

```python
# 모아봤는데 샘플수가 부족했음
# 그래서 추가로 68개의 샘플을 더해줌 (kick, hihat, tom 으로 추가)

import pandas as pd

# 전체 정보를 가지고 있는 엑셀 파일
all_info_file = "/Users/butterflyeffectleedea/desktop/git/Aiffel/aiffelton4.xlsx"
# 현재 232개의 샘플이 있는 엑셀 파일
current_file = "/Users/butterflyeffectleedea/desktop/git/Aiffel/aiffelton300.xlsx"

# 파일 불러오기
all_info_df = pd.read_excel(all_info_file)
current_df = pd.read_excel(current_file)

# 필요한 추가 샘플 수 설정
num_additional_samples = 68
additional_sc_values = ['kick', 'hihat', 'tom']

additional_samples = []

# 이미 선택된 샘플 제외
remaining_df = all_info_df[~all_info_df['filename'].isin(current_df['filename'])]

for sc in additional_sc_values:
    sc_df = remaining_df[remaining_df['sc'] == sc]

    # 추가 샘플 선택
    sample_size = min(len(sc_df), num_additional_samples // len(additional_sc_values))
    additional_samples.append(sc_df.sample(n=sample_size, replace=False, random_state=42))

# 추가 샘플 DataFrame 생성
additional_samples_df = pd.concat(additional_samples)

# 기존 결과와 추가 샘플 합치기
final_result_df = pd.concat([current_df, additional_samples_df])

# 결과를 새로운 Excel 파일로 저장
final_result_df.to_excel("/Users/butterflyeffectleedea/desktop/git/Aiffel/aiffelton300jjinmak.xlsx", index=False)

```

```python
# csv에서 fine-tuning용 데이터 = kick, snare, clap, hihat, cymbal, tom 을 제외하고 다른 instrument 는 제거

import pandas as pd

# CSV 파일 읽기
file_name = "output5.csv"
df = pd.read_csv(file_name)

# 허용되는 악기 목록
allowed_instruments = ['kick', 'snare', 'hihat', 'cymbal', 'tom', 'clap']

# instrument 열에서 허용되는 악기만 포함하는 행만 남김
df = df[df['Instrument'].isin(allowed_instruments)]

# 결과를 새로운 CSV 파일에 저장
df.to_csv("filtered_output5.csv", index=False)

print("악기 필터링이 완료되었습니다. 결과가 filtered_output.csv 파일에 저장되었습니다.")
```

```python
import os
import shutil
import pandas as pd

# 필터링된 CSV 파일 읽기
csv_file = "filtered_output5.csv"
df = pd.read_csv(csv_file)

# 원본 WAV 파일 경로
src_path = "/Users/butterflyeffectleedea/Downloads/preprocessdrum"

# 목적지 경로
dest_path = "/Users/butterflyeffectleedea/desktop/git/Aiffel/filtered_sounds"
dest_folder = "filtered_sounds"
dest_folder_path = os.path.join(dest_path, dest_folder)

# 목적지 폴더가 없으면 생성
if not os.path.exists(dest_folder_path):
    os.makedirs(dest_folder_path)

# 필터링된 파일 이름 목록 가져오기
filtered_filenames = df['File_Name'].tolist()

# 파일 복사
for filename in filtered_filenames:
    src_file_path = os.path.join(src_path, filename)
    dest_file_path = os.path.join(dest_folder_path, filename)
    
    if os.path.exists(src_file_path):
        shutil.copy2(src_file_path, dest_file_path)
    else:
        print(f"{filename} 파일을 찾을 수 없습니다.")

print("필터링된 파일이 새로운 폴더에 복사되었습니다.")

```
